```python
#STEP 2 - Import library
import pandas as pd
import numpy as np
from scipy.stats.mstats import winsorize
from google.colab import drive

#STEP 3 - Import dataset
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/housing_dirty.csv')
df.head()
```

### Remove Duplication

```python
#STEP 5 - Remove duplicate
print('Shape awal:', df.shape)
df.drop_duplicates(inplace=True)
print('Setelah hapus duplikat:', df.shape)
```

### Normalization

```python
#STEP 6 - Normalisasi
df['kota'] = df['kota'].str.strip().str.title()
df['kondisi'] = df['kondisi'].str.strip().str.lower()
df.head(5)
```

### Imputsasi Missing Values

```python
#STEP 7 - Imputasi Missing Values
df['luas_m2'] = df['luas_m2'].fillna(df['luas_m2'].median())
df['harga_juta'] = df['harga_juta'].fillna(df['harga_juta'].median())
df['kamar'] = df['kamar'].fillna(df['kamar'].median())
df.head(5)
```

### Handle Outlier

```python
#STEP 8 - Tangani Outlier
original_df = df.copy()

for col in ['harga_juta', 'luas_m2', 'tahun_bangun']:
  Q1, Q3 = df[col].quantile([0.25, 0.75])
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR

  # Identify outliers before clipping
  outlier_mask = (df[col] < lower_bound) | (df[col] > upper_bound)
  num_outliers_before = outlier_mask.sum()

  df[col] = df[col].clip(lower_bound, upper_bound)

  # Calculate changed values after clipping
  num_changed_values = (original_df[col] != df[col]).sum()
  print(f"Column '{col}': {num_changed_values} values were adjusted for outliers.")
```

### Validation

```python
#STEP 9 — Validasi
assert df.isnull().sum().sum() == 0, 'Masih ada missing!'
assert df.duplicated().sum() == 0, 'Masih ada duplikat!'
print('Shape akhir:', df.shape)
```

### Export

```python
#STEP 10 - Export
df.to_csv('housing_clean.csv', index=False)
print('Dataset bersih tersimpan!')
```
