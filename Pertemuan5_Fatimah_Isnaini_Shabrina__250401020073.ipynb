# Practice 5 - Visualization

## Import Library & Load Dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# style visual
sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (9,5)

# load dataset
tips = sns.load_dataset("tips")

# tampilkan data
tips.head()
```

## 📈 Grafik 1 — Rata-rata Total Bill per Hari
Visualisasi ini digunakan untuk melihat rata-rata pengeluaran pelanggan pada setiap hari.

```python
plt.figure(figsize=(9,5))

ax = sns.barplot(
    data=tips,
    x="day",
    y="total_bill",
    estimator='mean',
    palette="viridis"
)

# menampilkan angka di atas bar
for i in ax.containers:
    ax.bar_label(i, fmt='%.1f')

plt.title("Average Total Bill per Day", fontsize=14, weight='bold')
plt.xlabel("Day")
plt.ylabel("Average Total Bill")

plt.show()
```

### 📖 Analisis

**What?**  
Grafik menunjukkan rata-rata total bill pelanggan pada setiap hari. Hari Minggu memiliki rata-rata total bill tertinggi sekitar 21 dolar, sedangkan Jumat memiliki rata-rata paling rendah sekitar 17 dolar.

**So what?**  
Hal ini menunjukkan bahwa pelanggan cenderung mengeluarkan uang lebih banyak pada akhir pekan. Informasi ini penting bagi restoran untuk menentukan strategi operasional dan persediaan makanan.

**Now what?**  
Restoran dapat meningkatkan jumlah staf dan stok makanan pada akhir pekan untuk mengantisipasi peningkatan transaksi pelanggan.

## 📊 Grafik 2 — Distribusi Total Bill
Histogram digunakan untuk melihat persebaran total pembayaran pelanggan.

```python
plt.figure(figsize=(9,5))

sns.histplot(
    data=tips,
    x="total_bill",
    hue="time",
    bins=20,
    kde=True,
    color="skyblue"
)

plt.title("Distribution of Total Bill", fontsize=14, weight='bold')
plt.xlabel("Total Bill")
plt.ylabel("Frequency")

plt.show()
```

### 📖 Analisis

**What?**  
Sebagian besar transaksi pelanggan berada pada rentang 10–20 dolar. Transaksi dengan nilai di atas 40 dolar jumlahnya sangat sedikit.

**So what?**  
Distribusi ini menunjukkan bahwa mayoritas pelanggan melakukan transaksi dalam jumlah menengah sehingga restoran dapat menyesuaikan strategi harga menu.

**Now what?**  
Analisis lanjutan dapat dilakukan untuk mengetahui menu yang paling banyak berkontribusi terhadap transaksi besar.

## 📦 Grafik 3 — Perbandingan Tip Berdasarkan Status Perokok
Boxplot digunakan untuk melihat distribusi dan outlier pada data tip.

```python
plt.figure(figsize=(9,5))

sns.boxplot(
    data=tips,
    x="smoker",
    y="tip",
    palette="Set2"
)

plt.title("Tip Distribution by Smoker Status", fontsize=14, weight='bold')
plt.xlabel("Smoker Status")
plt.ylabel("Tip")

plt.show()
```

### 📖 Analisis

**What?**  
Distribusi tip pelanggan perokok dan non-perokok terlihat cukup mirip. Namun, kelompok perokok memiliki beberapa outlier dengan nilai tip tinggi.

**So what?**  
Hal ini menunjukkan bahwa status perokok tidak terlalu memengaruhi median tip, tetapi kelompok perokok memiliki variasi pengeluaran yang lebih besar.

**Now what?**  
Dapat dilakukan uji statistik lebih lanjut untuk mengetahui apakah terdapat perbedaan signifikan antara kedua kelompok.

## 🔍 Grafik 4 — Hubungan Total Bill dan Tip
Scatter plot digunakan untuk melihat hubungan antara total pembayaran dan tip pelanggan.

```python
plt.figure(figsize=(9,5))

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="time",
    palette="coolwarm",
    s=100
)

# Add regression line
sns.regplot(
    data=tips,
    x="total_bill",
    y="tip",
    scatter=False,
    color='black'
)

plt.title("Relationship Between Total Bill and Tip", fontsize=14, weight='bold')
plt.xlabel("Total Bill")
plt.ylabel("Tip")

plt.show()
```

### 📖 Analisis

**What?**  
Grafik menunjukkan hubungan positif antara total bill dan tip. Semakin besar total bill pelanggan, maka tip yang diberikan juga cenderung meningkat.

**So what?**  
Hubungan ini menunjukkan bahwa pelanggan dengan transaksi besar biasanya memberikan tip lebih tinggi, sehingga dapat menjadi indikator perilaku pelanggan restoran.

**Now what?**  
Analisis regresi dapat dilakukan untuk mengukur kekuatan hubungan antara total bill dan tip.

## Dashboard

```python
# ============================================================
# DASHBOARD VISUALISASI DATASET TIPS
# Pertemuan 5 - Visualisasi Data
# ============================================================

# ── 1. IMPORT LIBRARY ───────────────────────────────────────
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# ── 2. STYLE VISUAL ─────────────────────────────────────────
sns.set_style("whitegrid")
sns.set_context("talk")

# ── 3. LOAD DATASET ─────────────────────────────────────────
tips = sns.load_dataset('tips')

# informasi dataset
print("Jumlah Baris dan Kolom :", tips.shape)
print("\nInformasi Dataset:")
print(tips.info())

print("\nStatistik Deskriptif:")
print(tips.describe().round(2))

# ── 4. SETUP DASHBOARD ──────────────────────────────────────
fig = plt.figure(figsize=(18, 12))

fig.suptitle(
    '📊 Dashboard Analisis Dataset Tips',
    fontsize=22,
    fontweight='bold',
    y=0.98
)

# layout 2x2
gs = gridspec.GridSpec(
    2, 2,
    figure=fig,
    hspace=0.35,
    wspace=0.25
)

ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[1, 0])
ax4 = fig.add_subplot(gs[1, 1])

# ============================================================
# GRAFIK 1 — BAR CHART
# ============================================================

avg_by_day = tips.groupby('day')['total_bill'].mean()

bars = ax1.bar(
    avg_by_day.index,
    avg_by_day.values,
    color=sns.color_palette("viridis", 4),
    edgecolor='black'
)

# label angka di atas bar
for bar in bars:
    height = bar.get_height()
    ax1.text(
        bar.get_x() + bar.get_width()/2,
        height + 0.3,
        f'{height:.1f}',
        ha='center',
        fontsize=10
    )

ax1.set_title(
    'Average Total Bill per Day',
    fontsize=14,
    fontweight='bold'
)

ax1.set_xlabel('Day')
ax1.set_ylabel('Average Total Bill ($)')
ax1.spines[['top', 'right']].set_visible(False)

# ============================================================
# GRAFIK 2 — HISTOGRAM
# ============================================================

sns.histplot(
    data=tips,
    x='total_bill',
    hue='time',
    bins=20,
    kde=True,
    color='skyblue',
    ax=ax2
)

ax2.set_title(
    'Distribution of Total Bill',
    fontsize=14,
    fontweight='bold'
)

ax2.set_xlabel('Total Bill ($)')
ax2.set_ylabel('Frequency')
ax2.spines[['top', 'right']].set_visible(False)

# ============================================================
# GRAFIK 3 — BOXPLOT
# ============================================================

sns.boxplot(
    data=tips,
    x='smoker',
    y='tip',
    palette='Set2',
    ax=ax3
)

ax3.set_title(
    'Tip Distribution by Smoker Status',
    fontsize=14,
    fontweight='bold'
)

ax3.set_xlabel('Smoker Status')
ax3.set_ylabel('Tip ($)')
ax3.spines[['top', 'right']].set_visible(False)

# ============================================================
# GRAFIK 4 — SCATTER PLOT
# ============================================================

sns.scatterplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='time',
    palette='coolwarm',
    s=100,
    alpha=0.8,
    ax=ax4
)

# garis regresi
sns.regplot(
    data=tips,
    x='total_bill',
    y='tip',
    scatter=False,
    color='black',
    ax=ax4
)

ax4.set_title(
    'Relationship Between Total Bill and Tip',
    fontsize=14,
    fontweight='bold'
)

ax4.set_xlabel('Total Bill ($)')
ax4.set_ylabel('Tip ($)')
ax4.spines[['top', 'right']].set_visible(False)

# ============================================================
# FOOTNOTE
# ============================================================

fig.text(
    0.5,
    0.01,
    'Data Source: Seaborn Tips Dataset',
    ha='center',
    fontsize=11,
    color='gray'
)

# ── 5. SIMPAN DASHBOARD ────────────────────────────────────
plt.savefig(
    'dashboard_tips.png',
    dpi=300,
    bbox_inches='tight'
)

plt.show()

print("✅ Dashboard berhasil disimpan sebagai 'dashboard_tips.png'")
```
